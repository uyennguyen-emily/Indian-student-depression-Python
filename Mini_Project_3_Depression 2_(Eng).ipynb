```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_india = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MINI PROJECT_PYTHON/1 - India_Economic_Growth_Historical_Data.csv')
```

```python
df_india['GDP'] = df_india['GDP'].str.replace('$','').str.replace('B','').str.replace(',','').astype(float)
df_india['Per_Capita'] = df_india['Per_Capita'].str.replace('$','').str.replace(',','').astype(float)
df_india['Annual_percent_geowth'] = df_india['Annual_percent_geowth'].str.replace('%','').astype(float)
df_india
```

```python
# GDP
df_india = df_india[df_india['year'] > 2010]
fig = plt.figure(figsize=(15,7))
gdp_chart = sns.lineplot(data=df_india, x='year', y='GDP', color='orange', marker='o', markersize=8, linewidth=2, markeredgewidth=0)

for i, row in df_india.iterrows():
    x = row['year']
    y = row['GDP']
    plt.text(x, y + 25, f'{y}', ha='center', va='bottom', fontsize=12, fontweight='semibold')

sns.despine()
gdp_chart.spines['left'].set_visible(False)
gdp_chart.spines['bottom'].set_visible(False)
plt.legend(frameon=False, loc='upper left')
plt.grid(axis='y', color='gray', linestyle='--', alpha=0.3, zorder=0)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(integer=True))
plt.gca().tick_params(left=False, bottom=False)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.gca().set_facecolor('none')
fig.patch.set_alpha(0)

plt.title('2010-2022: Indian GDP shows an upward trend\n(unit: USD, billion)', fontsize=20, fontweight='bold', loc='left', pad=20)
plt.xlabel('', fontsize=14, labelpad=20, fontweight='semibold')
plt.ylabel('')
```

```python
# Annual Percent Growth
df_india = df_india[df_india['year'] > 2010]
fig = plt.figure(figsize=(15,7))
gdp_chart = sns.lineplot(data=df_india, x='year', y='Annual_percent_geowth', color='orange', marker='o', markersize=8, linewidth=2, markeredgewidth=0)

for i, row in df_india.iterrows():
    x = row['year']
    y = row['Annual_percent_geowth']
    plt.text(x, y+0.2, f'{y}', ha='center', va='bottom', fontsize=12, fontweight='semibold')

sns.despine()
gdp_chart.spines['left'].set_visible(False)
gdp_chart.spines['bottom'].set_visible(False)
plt.legend(frameon=False, loc='upper left')
plt.grid(axis='y', color='gray', linestyle='--', alpha=0.3, zorder=0)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(integer=True))
plt.gca().tick_params(left=False, bottom=False)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.gca().set_facecolor('none')
fig.patch.set_alpha(0)

plt.title('2010-2022: Annual growth rate is unstable', fontsize=20, fontweight='bold', loc='left', pad=20)
plt.xlabel('', fontsize=14, labelpad=20, fontweight='semibold')
plt.ylabel('')
```

```python
df_ppl = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MINI PROJECT_PYTHON/World Population by country 2024.csv')
df_ppl
```

```python
# Annual Percent Growth
df_ppl = df_ppl[['Country', 'Population 2024']].head(10)
fig = plt.figure(figsize=(15,7))
gdp_chart = sns.barplot(data=df_ppl, x='Country', y='Population 2024', palette='Oranges_r')

for a in gdp_chart.patches:
  if a.get_height() > 0:
    gdp_chart.annotate(f'{round(float(a.get_height() / 1000000),2)}',
                        (a.get_x() + a.get_width() / 2, a.get_height()+1000),
                        ha='center', va='bottom', fontsize=12, fontweight='semibold')
  a.set_zorder(3)

sns.despine()
plt.legend(frameon=False)
gdp_chart.spines['left'].set_visible(False)
gdp_chart.spines['bottom'].set_visible(False)
plt.gca().tick_params(left=False, bottom=False, color='darkgray')
plt.xticks(fontsize=11)
plt.yticks([])
plt.gca().set_facecolor('none')
fig.patch.set_alpha(0)
plt.grid(axis='y', color='gray', linestyle='--', alpha=0.3, zorder=0)

plt.title('India has the largest population in the world\n(unit: million people)', fontsize=16, fontweight='semibold', loc='left', pad=20)
plt.xlabel('', fontsize=14, labelpad=20, fontweight='semibold')
plt.ylabel('')
```
