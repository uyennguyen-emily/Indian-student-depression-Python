```python
# Dropna
df.dropna(inplace=True)

# Change datatype
df['Age'] = df['Age'].astype(int)
df['Financial Stress'] = df['Financial Stress'].astype(int)
df['Academic Pressure'] = df['Academic Pressure'].astype(int)
df['Work Pressure'] = df['Work Pressure'].astype(int)
df['Study Satisfaction'] = df['Study Satisfaction'].astype(int)
df['Job Satisfaction'] = df['Job Satisfaction'].astype(int)
df['Work/Study Hours'] = df['Work/Study Hours'].astype(int)


df['GenderID'] = df['Gender'].apply(lambda x: 1 if x == 'Male' else 0)
df['Suicidal Code'] = df['Have you ever had suicidal thoughts ?'].apply(lambda x: 1 if x == 'Yes' else 0)
df['Family Illness Code'] = df['Family History of Mental Illness'].apply(lambda x: 1 if x == 'Yes' else 0)

# Interpret encoded columns
def degree_eng(Degree):
  if Degree == 'B.Pharm':
    return 'Bachelor of Pharmacy'
  elif Degree == 'BSc':
    return 'Bachelor of Science'
  elif Degree == 'BA':
    return 'Bachelor of Arts'
  elif Degree == 'BCA':
    return 'Bachelor of Computer Applications'
  elif Degree == 'M.Tech':
    return 'Master of Technology'
  elif Degree == 'PhD':
    return 'Doctor of Philosophy'
  elif Degree == 'Class 12':
    return 'Secondary School Certificate'
  elif Degree == 'B.Ed':
    return 'Bachelor of Education'
  elif Degree == 'LLB':
    return 'Bachelor of Laws'
  elif Degree == 'BE':
    return 'Bachelor of Engineering'
  elif Degree == 'M.Ed':
    return 'Master of Education'
  elif Degree == 'MSc':
    return 'Master of Science'
  elif Degree == 'BHM':
    return 'Bachelor of Hotel Management'
  elif Degree == 'M.Pharm':
    return 'Master of Pharmacy'
  elif Degree == 'MCA':
    return 'Master of Computer Applications'
  elif Degree == 'MA':
    return 'Master of Arts'
  elif Degree == 'B.Com':
    return 'Bachelor of Commerce'
  elif Degree == 'MD':
    return 'Doctor of Medicine'
  elif Degree == 'MBA':
    return 'Master of Business Administration'
  elif Degree == 'MBBS':
    return 'Bachelor of Medicine, Bachelor of Surgery'
  elif Degree == 'M.Com':
    return 'Master of Commerce'
  elif Degree == 'B.Arch':
    return 'Bachelor of Architecture'
  elif Degree == 'LLM':
    return 'Master of Laws'
  elif Degree == 'B.Tech':
    return 'Bachelor of Technology'
  elif Degree == 'BBA':
    return 'Bachelor of Business Administration'
  elif Degree == 'ME':
    return 'Master of Engineering'
  elif Degree == 'MHM':
    return 'Master of Hotel Management'
  else:
    return 'Other Degrees'


df['Degree (Eng)'] = df['Degree'].apply(degree_eng)


# Interpret encoded columns using np.where()
df['Degree (Vie)'] = np.where(df['Degree'] == 'B.Pharm', 'Cử nhân Dược',
                      np.where(df['Degree'] == 'BSc', 'Cử nhân Khoa học',
                      np.where(df['Degree'] == 'BA', 'Cử nhân Nghệ thuật',
                      np.where(df['Degree'] == 'BCA', 'Cử nhân Ứng dụng Máy tính',
                      np.where(df['Degree'] == 'M.Tech', 'Thạc sĩ Công nghệ',
                      np.where(df['Degree'] == 'PhD', 'Tiến sĩ Triết học',
                      np.where(df['Degree'] == 'Class 12', 'Trung học Phổ thông',
                      np.where(df['Degree'] == 'B.Ed', 'Cử nhân Giáo dục',
                      np.where(df['Degree'] == 'LLB', 'Cử nhân Luật',
                      np.where(df['Degree'] == 'BE', 'Cử nhân Kỹ thuật',
                      np.where(df['Degree'] == 'M.Ed', 'Thạc sĩ Giáo dục',
                      np.where(df['Degree'] == 'MSc', 'Thạc sĩ Khoa học',
                      np.where(df['Degree'] == 'BHM', 'Cử nhân Quản trị Khách sạn',
                      np.where(df['Degree'] == 'M.Pharm', 'Thạc sĩ Dược',
                      np.where(df['Degree'] == 'MCA', 'Thạc sĩ Ứng dụng Máy tính',
                      np.where(df['Degree'] == 'MA', 'Thạc sĩ Nghệ thuật',
                      np.where(df['Degree'] == 'B.Com', 'Cử nhân Thương mại',
                      np.where(df['Degree'] == 'MD', 'Tiến sĩ Y khoa',
                      np.where(df['Degree'] == 'MBA', 'Thạc sĩ Quản trị Kinh doanh',
                      np.where(df['Degree'] == 'MBBS', 'Cử nhân Y khoa, Cử nhân Phẫu thuật',
                      np.where(df['Degree'] == 'M.Com', 'Thạc sĩ Thương mại',
                      np.where(df['Degree'] == 'B.Arch', 'Cử nhân Kiến trúc',
                      np.where(df['Degree'] == 'LLM', 'Thạc sĩ Luật',
                      np.where(df['Degree'] == 'B.Tech', 'Cử nhân Công nghệ',
                      np.where(df['Degree'] == 'BBA', 'Cử nhân Quản trị Kinh doanh',
                      np.where(df['Degree'] == 'ME', 'Thạc sĩ Kỹ thuật',
                      np.where(df['Degree'] == 'MHM', 'Thạc sĩ Quản trị Khách sạn', 'Khác'
                      )))))))))))))))))))))))))))

df['Degree Style'] = np.where (df['Degree (Eng)'].str.contains('Bachelor', na=False), 'Bachelor',
                              np.where(df['Degree (Eng)'].str.contains('Master', na=False), 'Master',
                              np.where(df['Degree (Eng)'].str.contains('Doctor', na=False), 'Doctor',
                              np.where(df['Degree (Eng)'].str.contains('Secondary School Certificate', na=False), \
                                       'Secondary School', df['Degree (Eng)']
                              ))))


df['Depression Explain'] = np.where(df['Depression'] == 0, 'No Depression', 'Depression')

# Age group
bins = [15, 20, 25, 30, 35, 40, 45, 50, 55, 100]
labels = ['16-20', '21-25', '26-30', '31-35', '36-40', '41-45', '46-50', '51-55', 'Over 55']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)
```

```python
# AGE GROUP & DEPRESSION
bins = [18, 20, 25, 30, 35, 40, 45, 50, 55, 60]
labels = ['16-20', '21-25', '26-30', '31-35', '36-40', '41-45', '46-50', '51-55', 'Over 55']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

df_age = df.groupby(['Age Group','Depression Explain']).size().reset_index(name='Quantity')
age_num = df['Age Group'].value_counts().reset_index(name='Total')
df_age = df_age.merge(age_num, on='Age Group')
df_age['Percent'] = round((df_age['Quantity'] / df_age['Total']) * 100, 2)

df_age = df_age[['Age Group', 'Depression Explain', 'Quantity', 'Total', 'Percent']]
df_age

fig = plt.figure(figsize=(15,10))
age_depbar = sns.lineplot(data=df_age, x='Age Group', y='Percent', hue='Depression Explain', palette='Oranges_r', linewidth=2, marker='o', markersize=10, markeredgewidth=0)

for age, percent in df_age.groupby('Depression Explain'):
  for a in range(len(percent)):
    x = a
    y = percent['Percent'].iloc[a]
    plt.text(a, y+1.5, f'{y}%', ha='center', va='bottom', fontsize=12, fontweight='semibold')

sns.despine()
age_depbar.spines['left'].set_visible(False)
age_depbar.spines['bottom'].set_visible(False)
plt.legend(frameon=False)
plt.grid(axis='y', color='gray', linestyle='--', alpha=0.3, zorder=0)
plt.grid(axis='x', which='both', linestyle='-', linewidth=0)
plt.gca().tick_params(left=False, bottom=False)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.gca().set_facecolor('none')
fig.patch.set_alpha(0)

plt.title('The risk of depression tends to decrease with age', fontsize=20, fontweight='semibold', loc='left', pad=20)
plt.xlabel('Age group', fontsize=14, labelpad=20, fontweight='semibold')
plt.ylabel('')
```
